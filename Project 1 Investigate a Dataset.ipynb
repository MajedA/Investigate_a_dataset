### Research Question 2:<br> 
### <pre>Best Genre.

<pre> 
In order to find the most popular movie genre, we will have to seperate each of the genres into it is own column so that we can get to our desired results.
</pre>

```python
def separate(data):
    return data.str[0:].str.split('|', expand = True)

genres = separate(movies_df['genres'])
genres['id'] = movies_df['id']
```

```python
def melt(df, col_name):
    return pd.melt(df, id_vars = 'id',value_name = col_name).drop(labels = 'variable',axis = 1)
genres = melt(genres, 'genres')
```

***
<pre>
After cleaning up the cluster. 
Let's expolre what are most requested genre.
</pre>

```python
plt.figure(figsize = (20,8), dpi = 150);
sns.countplot(genres['genres'],order = genres['genres'].value_counts().index, color='gold');
plt.ylabel("Movies",fontsize = 30);
plt.xlabel("Genres", fontsize = 40);
plt.xticks(rotation = 80, fontsize = 12);
plt.yticks(fontsize = 15);
```

<pre>
It seems that most demanded genres are Drama, then Comedy, then Thrillers and Action, which all scores above 2000.
